```python
# 각 데이터프레임의 인덱스를 리셋한 후 새로운 변수로 저장합니다.
ESM_result2 = ESM_result.reset_index()
ESS_result2 = ESS_result.reset_index()
ESN_result2 = ESN_result.reset_index()
MLPM_result2 = MLPM_result.reset_index()
MLPS_result2 = MLPS_result.reset_index()
MLPN_result2 = MLPN_result.reset_index()
SVRS_result2 = SVRS_result.reset_index()
SVRM_result2 = SVRM_result.reset_index()
SVRN_result2 = SVRN_result.reset_index()
LGM_result2 = LGM_result.reset_index()
LGS_result2 = LGS_result.reset_index()
LGN_result2 = LGN_result.reset_index()
GBM_result2 = GBM_result.reset_index()
GBS_result2 = GBS_result.reset_index()
GBN_result2 = GBN_result.reset_index()
RFM_result2 = RFM_result.reset_index()
RFS_result2 = RFS_result.reset_index()
RFN_result2 = RFN_result.reset_index()
LRM_result2 = LRM_result.reset_index()
LRS_result2 = LRS_result.reset_index()
LRN_result2 = LRN_result.reset_index()
```

```python
import pandas as pd

# 데이터프레임 목록 생성
df_list = [
    ESM_result2, ESS_result2, ESN_result2,
    MLPM_result2, MLPS_result2, MLPN_result2,
    SVRS_result2, SVRM_result2, SVRN_result2,
    LGM_result2, LGS_result2, LGN_result2,
    GBM_result2, GBS_result2, GBN_result2,
    RFM_result2, RFS_result2, RFN_result2,
    LRM_result2, LRS_result2, LRN_result2
]

# 데이터프레임 이름 목록 생성
df_names = [
    'ESM', 'ESS', 'ESN',
    'MLPM', 'MLPS', 'MLPN',
    'SVRS', 'SVRM', 'SVRN',
    'LGM', 'LGS', 'LGN',
    'GBM', 'GBS', 'GBN',
    'RFM', 'RFS', 'RFN',
    'LRM', 'LRS', 'LRN'
]

# 각 데이터프레임에 대해 Scaler 값을 업데이트
for df, name in zip(df_list, df_names):
    suffix = name[-1]  # 이름에서 마지막 문자를 추출
    if suffix == 'S':
        scaler_value = 'Standard'
    elif suffix == 'M':
        scaler_value = 'MinMax'
    elif suffix == 'N':
        scaler_value = 'Default'
    else:
        continue  # 해당하지 않으면 건너뜀
    
    # Scaler 컬럼 값을 업데이트
    if 'Scaler' in df.columns:
        df['Scaler'] = scaler_value
    else:
        print(f"DataFrame {name} does not have a 'Scaler' column.")
```

```python
all_results = [
    ESM_result2, ESS_result2, ESN_result2,
    MLPM_result2, MLPS_result2, MLPN_result2,
    SVRS_result2, SVRM_result2, SVRN_result2,
    LGM_result2, LGS_result2, LGN_result2,
    GBM_result2, GBS_result2, GBN_result2,
    RFM_result2, RFS_result2, RFN_result2,
    LRM_result2, LRS_result2, LRN_result2
]

# 모든 데이터프레임을 행 방향으로 병합
df_combined = pd.concat(all_results, ignore_index=True)
```

```python
col = ['Shift', 'Model', 'Scaler', 'index', 'train_r2', 'test_r2', 'train_mse', 'test_mse','train_mae_ratio', 'test_mae_ratio', 'test_mae_ratio_max','test_mae_ratio_min', 'test_mae_ratio_mean', 'train_mae', 'test_mae', 'train_rmse', 'test_rmse']
df_combined = df_combined[col]
```

```python
df_combined.to_csv("single_category_shift4.csv", index=False)
```
