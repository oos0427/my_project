```python
def Data_Comparison(a, b, c):
    """a = 데이터A
       b = 데이터B
       c = 비교할컬럼
    """
    set_a = set(a[c])
    set_b = set(b[c])
    common_values = set_a & set_b  # A와 B에 모두 있는 값
    only_in_a = set_a - set_b  # A에만 있는 값
    only_in_b = set_b - set_a  # B에만 있는 값

    # 결과 출력
    print("모두 있는 값:")
    print(common_values)

    print("\nA에만 있는 값:")
    print(only_in_a)

    print("\nB에만 있는 값:")
    print(only_in_b)
```

```python
def Data_Comparison2(dict_to_compare, c):
    """
    dict_to_compare = {"sales" :sales, "store": store }
    """

    keys = list(dict_to_compare.keys())

    set_a = set(dict_to_compare[keys[0]][c])
    set_b = set(dict_to_compare[keys[1]][c])
    common_values = set_a & set_b  # A와 B에 모두 있는 값
    only_in_a = set_a - set_b  # A에만 있는 값
    only_in_b = set_b - set_a  # B에만 있는 값

    # 결과 출력
    print("모두 있는 값:")
    print(common_values)

    print(f"\n{keys[0]}에만 있는 값:")
    print(only_in_a)

    print(f"\n{keys[1]}에만 있는 값:")
    print(only_in_b)
```

```python
Data_Comparison2({"sales" :sales, "store": store }, '행정동')
```

```python
Data_Comparison2({"resident" : resident, "store": store }, '행정동')
```

```python
Data_Comparison2({"sales" : sales, "store": store }, '년분기')
```

```python
Data_Comparison2({"sales" : sales, "store": store }, '업종명')
```

<h1>상관관계분석 시운전</h1>

```python
sales["업종명"].unique()
```

```python
import pandas as pd
merged_data = pd.read_csv("Merged_Sales_data_v2.2.csv")
```

```python
c = merged_data[merged_data["업종명"] == "한식음식점"]
c.head()
import pandas as pd

# Step 1: 숫자형 데이터 선택
numeric_data = c.select_dtypes(include=['number'])

# Step 2: '행정동'과 '업종명' 원-핫 인코딩
one_hot_encoded = pd.get_dummies(c[['행정동', '업종명']], prefix=['행정동', '업종명'])

# Step 3: 숫자형 데이터와 원-핫 인코딩 결합
combined_data = pd.concat([numeric_data, one_hot_encoded], axis=1)

# Step 4: 상관관계 계산
correlation_matrix = combined_data.corr()

# Step 5: 데이터프레임으로 변환
correlation_df = pd.DataFrame(correlation_matrix)

# Step 6: 상관관계 결과 확인
print("상관관계 분석 결과 (데이터프레임 형식):")
print(correlation_df)
# '월매출' 열 기준으로 내림차순 정렬
sorted_correlation = correlation_df["월매출"].sort_values(ascending=False)
```

```python
# 출력
print("월매출과의 상관관계 (내림차순 정렬):")
print(sorted_correlation[1:11])
print(sorted_correlation[-10:])
```
