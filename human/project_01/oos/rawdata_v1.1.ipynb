수정한 내용 마체중 및 트랙상태

트랙상태 포화(18%) -> TR_CON : 포화, TR_WC : 18 <br>
마체중 311(+19) -> horse_weight : 311, weight_gain : 19
* ()의 경우 0으로 입력, 그외 디테일은 아래 코드 참조

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('source/horse_ecolumn.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
# TRACK 컬럼을 TR_STATE와 TR_WC로 분리
df[['TR_CON', 'TR_WC']] = df['TRACK'].str.extract(r'(\D+)\s\((\d+)%\)')

# NaN 값을 0으로 대체하고 TR_WC 컬럼을 int 타입으로 변환
df['TR_WC'] = df['TR_WC'].fillna(0).astype(int)

# 원래 TRACK 컬럼 삭제
df = df.drop(columns=['TRACK'])
```

```python
df.loc[:, ['TR_CON', 'TR_WC']]
```

```python
df.columns
```

```python
# '마체중'과 '증감'으로 분리
df['horse_weight'] = df['WG_HR'].str.extract(r'(\d+)').astype(float)  # 숫자 부분 추출
df['weight_gain'] = df['WG_HR'].str.extract(r'\(([-+]?\d+)\)').fillna(0).astype(float)  # 괄호 안의 숫자 추출, 빈 경우 0으로 대체
```
