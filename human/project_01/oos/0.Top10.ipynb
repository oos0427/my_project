<h1>전체기간 지역별 1등 많이한 10개 말</h1>
2019년 전에 경기도 포함되는 값(ORD1_CNT_T)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('source/raw_vol2.0.csv')
```

```python
df.head(1)
```

```python
#말의 이름으로 그룹화 이후 1등 많이한 횟수 반환 (2019년 이전 기록 포함)
ORD1_CNT_T = df.loc[df.groupby('HR_NAME')['ORD1_CNT_T'].idxmax()].sort_values(by='ORD1_CNT_T', ascending=False)
```

```python
ORD1_CNT_T.iloc[:3,:20]
```

```python
top_10_df = ORD1_CNT_T.head(10)
```

```python
top_10_df.iloc[:, :20]
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows의 경우
# 막대그래프 생성
plt.figure(figsize=(10, 6))
plt.bar(top_10_df['HR_NAME'], top_10_df['ORD1_CNT_T'])  # x축: 말 이름, y축: 1등 횟수
plt.xlabel('Horse Name')
plt.ylabel('1st Place Count')
plt.title('JEJU Top 10 Horses by 1st Place Count')
plt.xticks(rotation=45)  # 말 이름이 잘 보이도록 45도 회전
plt.show()
```

```python
ORD1_CNT_T_Busan = ORD1_CNT_T[ORD1_CNT_T['MEET'] == '부산경남']
```

```python
top_10_busan = ORD1_CNT_T_Busan.head(10)
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows의 경우
# 막대그래프 생성
plt.figure(figsize=(10, 6))
plt.bar(top_10_busan['HR_NAME'], top_10_busan['ORD1_CNT_T'])  # x축: 말 이름, y축: 1등 횟수
plt.xlabel('Horse Name')
plt.ylabel('1st Place Count')
plt.title('BUSAN Top 10 Horses by 1st Place Count')
plt.xticks(rotation=45)  # 말 이름이 잘 보이도록 45도 회전
plt.show()
```
