```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr, kendalltau
```

```python
df = pd.read_csv('source/raw_vol2.0.csv')
```

```python
df2 = df.loc[:, ['RC_DIST','RANK', 'ORD', 'AGE', 'horse_weight', 'weight_gain', 'RC_TIME', 'WG_BUDAM', 'JK_AGE', 'JK_CAREER', 'TR_CAREER']]
```

```python
df2['RANK'].unique()
```

```python
df2_name_sorted = [
    '제1등급', '제2등급', '제3등급', '제4등급', '제5등급', '제6등급',
    '한1등급', '한2등급', '한3등급', '한4등급',
    '국1등급', '국2등급', '국3등급', '국4등급', '국5등급', '국6등급'
]
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'


for rank in df2_name_sorted:
    #plt.figure(figsize=(12, 10))
    df3 = df2[df2['RANK'] == rank]
    df4 = df3.drop(columns='RANK')
    sns.heatmap(df4.iloc[:, 1:].corr().round(2), annot=True, cmap='coolwarm')
    plt.title(f'{rank} 의 상관관계')
    plt.show()
```

```python
# 숫자형 열만 선택
df2 = df.loc[:, ['RC_DIST','RANK', 'ORD', 'AGE', 'horse_weight', 'weight_gain', 'RC_TIME', 'WG_BUDAM', 'JK_AGE', 'JK_CAREER', 'TR_CAREER']]
numeric_df = df2.select_dtypes(include=['number'])

# 상관계수 히트맵 그리기
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr().round(2), annot=True, cmap='coolwarm')
plt.title('상관관계')
plt.show()
```
