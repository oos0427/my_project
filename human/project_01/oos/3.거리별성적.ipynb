```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr, kendalltau
```

```python
df = pd.read_csv('source/raw_vol2.0.csv')
```

```python
df['RC_DIST'].unique()
```

```python
rc_dist = [ 800,  900, 1000, 1110, 1200, 1400, 1700, 1610, 1300, 1600, 1800,
       2000, 2200, 1900]
```

```python
df = df[df['ORD'] == 1]
```

```python
df.groupby('RC_DIST')['MEET'].count()
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'

for dist in rc_dist:
    df2 = df[df['RC_DIST'] == dist]
    df2 = df2.groupby('HR_NAME')['MEET'].count().sort_values(ascending=False).reset_index()
    df3 = df2.iloc[:10, :]
    # 막대그래프 생성
    plt.figure(figsize=(10, 6))
    plt.bar(df3['HR_NAME'], df3['MEET'], color='skyblue')
    plt.xlabel('HR_NAME')  # x축 라벨
    plt.ylabel('count of 1rd')  # y축 라벨
    plt.title(f'{dist}M: Top 10 Horse Names by 1rd Count')  # 제목
    plt.xticks(rotation=45)  # x축 라벨 회전
    plt.tight_layout()  # 레이아웃 조정
    plt.show()
```
